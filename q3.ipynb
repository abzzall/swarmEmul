```python
directory = 'q3'
csv_file_name=os.path.join(directory, 'report.csv')
df= pandas.read_csv(csv_file_name, header=None, names=['id', 'sensor_range' , 'n1', 'n2', 'n3', 'N', 'lived_agents', 'leader_goal_distance', 't', 'w1', 'w2', 'w3'])
```

```python
def plt3d(i, df, w, isT, dir=directory, suffix=''):
    title='3d_'
    fig=plt.figure()
    ax = plt.axes(projection ='3d')
    if i==1:
        df_=df[df['w1']==w]
        x=df_['w2']
        y=df_['w3']
        title+='w1_'+str(w)
        ax.set_xlabel('w2', labelpad=20)
        ax.set_ylabel('w3', labelpad=20)
    elif i==2:

        df_=df[df['w2']==w]
        x=df_['w1']
        y=df_['w3']
        title+='w2_'+str(w)
        ax.set_xlabel('w1', labelpad=20)
        ax.set_ylabel('w3', labelpad=20)
    elif i==3:

        df_=df[df['w3']==w]
        x=df_['w1']
        y=df_['w2']
        title+='w3_'+str(w)
        ax.set_xlabel('w1', labelpad=20)
        ax.set_ylabel('w2', labelpad=20)
    if isT:
        z=df_.t
        title+="_t"
        ax.set_zlabel('t', labelpad=20)
    else:
        z=df_.lived_agents
        title+="_lived_agent"
        ax.set_zlabel('lived_agent', labelpad=20)
    print(w, title)
    ax.plot_trisurf(x, y, z, linewidth=0.2, antialiased=True)
    # ax.plot_surface(x, y, z, cmap ='viridis', edgecolor ='green')
    title+=suffix
    ax.set_title(title)
    fig.savefig(os.path.join(dir, title+'.png'))

```

```python
def plt2d(i, df, w1, w2, isT, dir=directory, suffix=''):
    title='2d_'
    fig=plt.figure()
    ax = plt.axes()
    if i==1:
        df_=df[df['w2']==w1]

        df_=df_[df_['w3']==w2]
        x=df_['w1']
        title+='w2_'+str(w1)+'_w3_'+str(w2)
        ax.set_xlabel('w1', labelpad=20)
    elif i==2:
        df_=df[df['w1']==w1]

        df_=df_[df_['w3']==w2]
        x=df_['w2']
        title+='w1_'+str(w1)+'_w3_'+str(w2)
        ax.set_xlabel('w2', labelpad=20)
    elif i==3:

        df_=df[df['w1']==w1]

        df_=df_[df_['w2']==w2]
        x=df_['w3']
        title+='w1_'+str(w1)+'_w2_'+str(w2)
        ax.set_xlabel('w3', labelpad=20)
    if isT:
        z=df_.t
        title+="_t"
        ax.set_ylabel('t', labelpad=20)
    else:
        z=df_.lived_agents
        title+="_lived_agent"
        ax.set_ylabel('lived_agent', labelpad=20)
    ax.plot(x, z)
    ax.set_title(title)
    title=title+suffix
    print(w, title)

    fig.savefig(os.path.join(dir, title+'.png'))
```
